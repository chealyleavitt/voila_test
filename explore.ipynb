```python
import pandas as pd
import numpy as np
import bokeh

import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
lc = pd.read_excel('data/41467_2021_27797_MOESM5_ESM.xlsx', sheet_name = 'Sheet1')

# drop rows with PACS== None

lc = lc.dropna(subset=["PACS"])

# Create new column called "group" where group is PACS YES, NO, CONTROL

lc['PACS Group']=["CONTROL" if x=='Healthy' else y for x,y in zip(lc.COVID,lc.PACS)]
```

```python
lc.columns
```

```python
sns.set_theme(style="ticks")
f1, ax1 = plt.subplots(figsize=(7, 5))
sns.despine(f1)

f1 = sns.histplot(lc[lc.Sampling_month=="ONE"],
                  x="Age", hue="PACS Group",
                  multiple="stack",
#                   palette=["gold", "lawngreen", "lightseagreen"],
                  palette="colorblind",
                  edgecolor=".3",
                  linewidth=.5,
                 )
f1.set_title("Age by Group")
plt.show()
```

```python
f2, ax2 = plt.subplots(figsize=(7, 5))
sns.despine(f2)
f2 = sns.histplot(data=lc[lc.Sampling_month=="ONE"],
                  multiple="stack",
                  x="Asthma",
                  hue="PACS Group",
                  palette="colorblind",
                  edgecolor=".3",
                  linewidth=.5,
)

plt.show()
```
